# Student Performance Analysis Project

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/PS01_Student_Performance.csv')

# Display the first 5 rows of the DataFrame
print("First 5 rows of the dataset:")
display(df.head())
```

## Data Visualization with Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style("whitegrid")

# Ensure 'Total_Score' and 'Average_Score' columns exist for plotting
# Recalculate these to make the cell robust against out-of-order execution
df['Total_Score'] = df[['Maths', 'Science', 'English']].sum(axis=1)
df['Average_Score'] = df[['Maths', 'Science', 'English']].mean(axis=1)

# Visualization 1: Distribution of Average Scores
plt.figure(figsize=(10, 6))
sns.histplot(df['Average_Score'], kde=True, bins=5)
plt.title('Distribution of Student Average Scores')
plt.xlabel('Average Score')
plt.ylabel('Number of Students')
plt.show()

# Visualization 2: Bar chart of average scores per subject
average_subject_scores = df[['Maths', 'Science', 'English']].mean()
plt.figure(figsize=(8, 5))
sns.barplot(x=average_subject_scores.index, y=average_subject_scores.values, hue=average_subject_scores.index, palette='viridis', legend=False)
plt.title('Average Scores per Subject')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.ylim(0, 100) # Assuming scores are out of 100
plt.show()

# Visualization 3: Bar chart of individual total scores
plt.figure(figsize=(12, 6))
sns.barplot(x='Name', y='Total_Score', data=df.sort_values('Total_Score', ascending=False), hue='Name', palette='plasma', legend=False)
plt.title('Total Scores Per Student')
plt.xlabel('Student Name')
plt.ylabel('Total Score (Maths + Science + English)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Summary of Insights

Key insights from the student performance analysis include:

1.  **Data Quality**: The dataset is robust and clean, with no missing values or duplicates.

2.  **Overall Performance Distribution**: Average scores show varied performance across students, with specific concentrations observed.
